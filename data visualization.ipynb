# relationship plot

```python
abi.relplot(x='total_bill',y='tip',data=tip,col="time",hue="day" ,style="day",kind="line")
```

```python
abi.relplot(x='total_bill',y='tip',data=tip,col="time",hue="day" ,style="day",kind="scatter")
```

```python
abi.relplot(x='total_bill',y='tip',data=tip,col="size",hue="day" ,style="day",kind="scatter")
```

```python
#categorical box plot
```

```python
abi.catplot(x='size',y='total_bill',kind='box',data=tip)
```

# regression  plot

```python
abi.regplot(x='total_bill',y='tip',data=tip)
```
